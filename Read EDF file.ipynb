```python
import pyedflib
import xml.etree.ElementTree
import numpy as np
import matplotlib.pyplot as plt
```

```python
edf = pyedflib.EdfReader("1.edf")
```

```python
def hypnogramFromXmlFile(xml_file_name):
    root = xml.etree.ElementTree.parse(xml_file_name).getroot()
    hypnogram = np.zeros((edf.file_duration // 30, 1), dtype=np.int)
    for i, event in enumerate(root.iter('SleepStage')):
        stage = int(event.text)
        hypnogram[i] = stage
    return hypnogram
```

```python
def annotationsFromXmlFile(xml_file_name, label):
    root = xml.etree.ElementTree.parse(xml_file_name).getroot()
    annotations = []
    for event in root.iter('ScoredEvent'):
        if event[0].text == label:
            annotations.append({"start": float(event[1].text), 
                                "duration": float(event[2].text)})
    return annotations
```

```python
print("File name: ", edf.file_name)
print("PSG duration (seconds): ", edf.file_duration)
print("Signals info:")
for i in range(1, edf.signals_in_file):
    # You can get the complete header with
    # edf.getSignalHeader(i)
    print("  # Index: ", i)
    print("  Label: ", edf.getSignalHeader(i)['label']) 
    print("  Rate: ", edf.getSignalHeader(i)['sample_rate'])
```
